```python
class Refuel:
    def __init__(self, N, E, file, is_main, primaryMinX, primaryMinY, primaryMaxX, primaryMaxY, initial_belief):
        self.N = N
        self.E = E
        self.file = file
        self.is_main = is_main
        self.primaryMinX = primaryMinX
        self.primaryMinY = primaryMinY
        self.primaryMaxX = primaryMaxX
        self.primaryMaxY = primaryMaxY
        self.obstacles, self.rfstations = self.ini_obstacles_rfstations()
        self.belief = initial_belief
    def ini_obstacles_rfstations(self):
        obstacles = set()
        rfstations = set()
        N, E = self.N, self.E
        if N == 6 and E == 8:
            obstacles.add((4,4))
            rfstations.add((0,0))
            rfstations.add((2,2))
            rfstations.add((3,3))
        elif N == 12 and E == 16:
            obstacles.add((4,4))
            obstacles.add((4,4))
            rfstations.add((0,0))
            rfstations.add((2,2))
            rfstations.add((3,3))
        obstacles = sorted(list(obstacles))
        rfstations = sorted(list(rfstations))
        return obstacles, rfstations
    
    def in_bound(self, x, y):
#         x, y = object[0], object[1]
        lx, ly, rx, ry = self.primaryMinX, self.primaryMinY, self.primaryMaxX, self.primaryMaxY
        borderSize = 2
        areaMinX = max(lx - borderSize, 0)
        areaMinY = max(ly - borderSize, 0)
        areaMaxX = min(rx + borderSize, self.N - 1)
        areaMaxY = min(ry + borderSize, self.N - 1)
        #return lx <= x <= rx and ly <= y <= ry #in primary bound
        return areaMinX <= x <= areaMaxX and areaMinY <= y <= areaMaxY 
        
    def plot_board(self):
        from matplotlib import pyplot as plt
        beliefState= self.belief
        N = self.N
        obstacles = self.obstacles
        rfstations = self.rfstations
        agent = None
        fig, ax = plt.subplots(figsize = (min(N,100),min(N,100)))
        #ax.xaxis.set_ticks_position('top')   #将X坐标轴移到上面
        ax.invert_yaxis()  
        #
        x = [i for i in range(N)]
        y = [i for i in range(N)]
        plt.xlim((0,N))
        #plt.ylim((0,N))
        ax.set_ylim(bottom=N, top=0)  # 此处将原点设置为左上角  
        plt.xticks([0.5 + i for i in range(N )], labels = [str(i) for i in range(N)])
        plt.yticks([0.5 + i for i in range(N)], labels = [str(i) for i in range(N)])
        plt.xlabel('X Axis')
        plt.ylabel('Y Axis')
        width = 1
        height =1 

        for left, bottom in obstacles:
            rect = plt.Rectangle((left, bottom), width, height, facecolor="red", alpha = 1)
            ax.add_patch(rect)

        for left, bottom in rfstations:
            rect = plt.Rectangle((left, bottom), width, height, facecolor="yellow", alpha = 1)
            ax.add_patch(rect)

        for left, bottom in beliefState:
            rect = plt.Rectangle((left, bottom), width, height, facecolor="blue", alpha = 0.5)
            ax.add_patch(rect)

        #plt.scatter(agent[0] + 0.5, agent[1] +0.5)
        for i in range(N):
            plt.hlines(i, 0, N , alpha = 0.5)
            plt.vlines(i, 0, N , alpha = 0.5)
        plt.grid(False)
        plt.savefig("./N{}E{}/refule_N{}E{}.jpg".format(self.N, self.E, self.N, self.E))  
        plt.show()     
        
    def print_primary_grid(self, file):
        lx, ly, rx, ry = self.primaryMinX, self.primaryMinY, self.primaryMaxX, self.primaryMaxY
        print("const int primaryMinX = {};".format(lx), file = file)
        print("const int primaryMinY = {};".format(ly), file = file)
        print("const int primaryMaxX = {};".format(rx), file = file)
        print("const int primaryMaxY = {};".format(ry), file = file)
        print("const int borderSize = 2;", file = file)
        print("const int areaMinX = max(primaryMinX - borderSize, axMIN);", file = file)
        print("const int areaMinY = max(primaryMinY - borderSize, ayMIN);", file = file)
        print("const int areaMaxX = min(primaryMaxX + borderSize, axMAX);", file = file)
        print("const int areaMaxY = min(primaryMaxY + borderSize, ayMAX);", file = file)
    
    def print_obstacles_rfstations(self, file):
        for i, (x, y) in enumerate(self.obstacles):
            if not self.in_bound(x, y):
                continue
            print("const int ob{}x = {};".format(i + 1, x), file=file)
            print("const int ob{}y = {};".format(i + 1, y), file=file)
        for i, (x, y) in enumerate(self.rfstations):
            #if not self.in_bound(x, y):
                #continue
            print("const int rf{}x = {};".format(i + 1, x), file=file)
            print("const int rf{}y = {};".format(i + 1, y), file=file)
    def print_model(self):  
        with open(self.file, 'w') as file:
            print("pomdp", file=file)
            print("observables", file=file)
            print("start", file=file)
            print("endobservables", file=file)
            print("", file=file)
            print("const int N = {};".format(self.N), file=file)
            print("const int ENERGY = {} ;".format(self.E), file=file)
            print("const int RESOLUTION = 2;", file=file)
            print("const int fuelCAP = ENERGY;", file=file)
            print("const int axMAX = N-1;", file=file)
            print("const int ayMAX = N-1;", file=file)
            print("const int axMIN = 0;", file=file)
            print("const int ayMIN = 0;", file=file)
            print("const double slippery = 0.3;", file=file)

            self.print_primary_grid(file)
            self.print_obstacles_rfstations(file)
            
            print("", file=file)
            print("formula distance = axMAX - ax + ayMAX - ay;", file=file)
            print("formula distrf1 = max(rf1x - ax, ax - rf1x) + max(rf1y - ay, ay - rf1y);", file=file)
            print("formula westenabled = ax != axMIN;", file=file)
            print("formula eastenabled = ax != axMAX;", file=file)
            print("formula northenabled = ay != ayMIN;", file=file)
            print("formula southenabled = ay != ayMAX;", file=file)
            print("observable \"cangonorth\" = northenabled;", file=file)
            print("observable \"cangosouth\" = southenabled;", file=file)
            print("observable \"cangowest\" = westenabled;", file=file)
            print("observable \"cangoeast\" = eastenabled;", file=file)
            

            if self.is_main:
                done = "formula done = start & ax = axMAX &  ay = ayMAX ;"
            else:
                done =  "formula done = start & (!crash) & ((ax = axMAX & ay = ayMAX) | "
                done += "ax = primaryMaxX + 1 | ax = primaryMaxX + 2 | ay = primaryMaxY + 1 | ay = primaryMaxY + 2 |"
                done += "ax = primaryMinX - 1 | ax = primaryMinX - 2 | ay = primaryMinY - 1 | ay = primaryMinY - 2 ) "
                #done += "& (distance >= fuel)"
                done += ";"
            print(done, file = file)
            
            print("observable \"amdone\" = done;", file=file)
            
            #self.print_formula()
            formula_crash = "formula crash ="
            crash_obstacle = []
            for i, (x, y) in enumerate(self.obstacles):
                if not self.in_bound(x, y):
                    continue
                crash_obstacle.append("(ax = ob{}x & ay = ob{}y)".format(i + 1, i + 1))
            if not crash_obstacle:
                formula_crash += " false;"
            else:
                formula_crash += "start & (" + " | ".join(crash_obstacle) + ");"
            print(formula_crash, file=file)
            
            formula_station = "formula atStation ="
            at_station = []
            for i, (x, y) in enumerate(self.rfstations):
                if not self.in_bound(x, y):
                    continue
                at_station.append("(ax = rf{}x & ay = rf{}y)".format(i + 1, i + 1))
            if not crash_obstacle:
                formula_station += " false;"
            else:
                formula_station += "start & (" +  " | ".join(at_station) + ");"
            print(formula_station, file=file)
            
            #print("formula atStation = (ax = rf1x & ay = rf1y) | (ax = rf2x & ay = rf2y) |  (ax = rf3x & ay = rf3y);", file=file)
            print("formula canRefuel = atStation  & fuel < fuelCAP;", file=file)
            
            print("observable \"hascrash\" = crash;", file=file)
            print("observable \"refuelAllowed\" = canRefuel;", file=file)
            print("observable \"fuelempty\" = fuel > 0;", file=file)
            print("observable \"fuelfull\" = fuel = ENERGY;", file=file)
            print("observable \"fuelmeter\" = floor(fuel/(ENERGY/RESOLUTION));", file=file)
            print("", file=file)
            
            print("module master", file=file)
            print("start : bool init false;", file=file)
            print("", file=file)
            print("[placement] !start -> (start\'=true);", file=file)
            print("[north] start & !done -> true;", file=file)
            print("[south] start  & !done -> true;", file=file)
            print("[east] start  & !done-> true;", file=file)
            print("[west] start & !done -> true;", file=file)
            print("[done] start & done -> true;", file=file)
            print("endmodule", file=file)
            print("", file=file)
            
            print("module tank", file=file)
            print("fuel : [0..fuelCAP] init fuelCAP;", file=file)
            print("", file=file)
            if not is_main:
                pre = "[placement] true -> "
                initial_fuel = []
                for fuel in range(self.E + 1):
                    initial_fuel.append("1/{}: (fuel\'={})".format(self.E + 1, fuel))
                pre += " + ".join(initial_fuel) + ";"
                print(pre, file=file)
            print("[refuel] canRefuel -> 1:(fuel\'=fuelCAP);", file=file)
            print("[north] fuel > 0 & !canRefuel -> 0.7:(fuel\'=fuel-1) + 0.3:(fuel\'=max(fuel-1,0));", file=file)
            print("[south] fuel > 0 & !canRefuel -> 0.7:(fuel\'=fuel-1)+ 0.3:(fuel\'=max(fuel-1,0));", file=file)
            print("[east] fuel > 0 & !canRefuel -> 0.7:(fuel\'=fuel-1)+ 0.3:(fuel\'=max(fuel-1,0));", file=file)
            print("[west] fuel > 0 & !canRefuel -> 0.7:(fuel\'=fuel-1)+ 0.3:(fuel\'=max(fuel-1,0));", file=file)
            print("[empty] fuel = 0 & !canRefuel -> 1:(fuel\'=0);", file=file)
            print("endmodule", file=file)
            print("", file=file)
            
            print("module rover", file=file)
            if self.is_main:
                print("ax : [axMIN..axMAX] init 0;", file=file)
                print("ay : [ayMIN..ayMAX] init 0;", file=file)
                print("[placement] true ->  1: (ax\'=areaMinX) & (ay\'=areaMinY); //+ 1/4: (ax\'=1) & (ay\'=1) + 1/4: (ax\'=2) & (ay\'=1) + 1/4: (ax\'=1) & (ay\'=3);", file=file)
            else:
                print("ax : [areaMinX..areaMaxX] init areaMinX;", file=file)
                print("ay : [areaMinY..areaMaxY] init areaMinY;", file=file)              
                #initial_place = ["1:(ax\'=primaryMinX+1) & (ay\'=primaryMinY)"]
                initial_place = ["1/4:(ax\'=primaryMinX) & (ay\'=primaryMinY)", "1/4:(ax\'=primaryMinX) & (ay\'=primaryMaxY)",
                         "1/4:(ax\'=primaryMaxX) & (ay\'=primaryMinY)", "1/4:(ax\'=primaryMaxX) & (ay\'=primaryMaxY)"]
                print("[placement] true ->" + " + ".join(initial_place) + ";", file = file)
            print("[west] westenabled -> (1-slippery): (ax\'=max(ax-1,axMIN)) + slippery: (ax\'=max(ax-2,axMIN));", file=file)
            print("[east] eastenabled -> (1-slippery): (ax\'=min(ax+1,axMAX)) + slippery: (ax\'=min(ax+2,axMAX));", file=file)
            print("[south]  southenabled -> (1-slippery): (ay\'=min(ay+1,ayMAX)) + slippery: (ay\'=min(ay+2,ayMAX));", file=file)
            print("[north]  northenabled -> (1-slippery): (ay\'=max(ay-1,ayMIN)) + slippery: (ay\'=max(ay-2,ayMIN));", file=file)
            print("endmodule", file=file)
            print("", file=file)
            
            print("rewards \"steps\"", file=file)
            print("[north] true : 1;", file=file)
            print("[south] true : 1;", file=file)
            print("[west] true : 1;", file=file)
            print("[east] true : 1;", file=file)
            print("endrewards", file=file)
            print("", file=file)
            print("rewards \"refuels\"", file=file)
            print("[refuel] true : 1;", file=file)
            print("endrewards", file=file)
            print("", file=file)
            print("rewards \"cost\"", file=file)
            print("[north] true : 1;", file=file)
            print("[south] true : 1;", file=file)
            print("[west] true : 1;", file=file)
            print("[east] true : 1;", file=file)
            print("[refuel] true : 3;", file=file)
            print("endrewards", file=file)
            print("", file=file)
            print("rewards \"reward\"", file=file)
            print("done : 10;", file=file)
            print("endrewards", file=file)
            print("label \"goal\" = done & (fuel >= distance | atStation | fuel >= distrf1);", file=file)
            print("label \"traps\" = crash;", file=file)
            print("label \"stationvisit\" = atStation;", file=file)
            print("label \"notbad\" =  !crash & (fuel > 0 | canRefuel);", file=file)
        file.close()
```
